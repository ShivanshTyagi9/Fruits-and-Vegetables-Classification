```python
import json
```

```python
# Load the saved history
with open('ResNet_history.json', 'r') as f:
    ResNet_history = json.load(f)

with open('EffNet_history.json', 'r') as f:
    EffNet_history = json.load(f)
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_model_comparison(history1, history2):
    plt.figure(figsize=(10, 6))
    
    # Plot ResNet
    plt.plot(history1['accuracy'], 'b-', label='Train Accuracy - ResNet')
    plt.plot(history1['val_accuracy'], 'orange', label='Validation Accuracy - ResNet')
    
    # Plot EfficientNet
    plt.plot(history2['accuracy'], 'g--', marker='x', label='Train Accuracy - EffNet')
    plt.plot(history2['val_accuracy'], 'r--', marker='x', label='Validation Accuracy - EffNet')
    
    plt.title('Comparison of Train and Validation Accuracy for Two Models')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(loc='best')
    plt.tight_layout()
    plt.show()

plot_model_comparison(ResNet_history, EffNet_history)
```
